- $X_0 = (0,0)$

```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_trayectorias_2D(traj_jacobi, traj_gs):
    # Convertir trayectorias a arrays 2 x k
    TJ = np.hstack(traj_jacobi)
    TG = np.hstack(traj_gs)

    plt.figure()

    # Jacobi
    plt.plot(TJ[0, :], TJ[1, :], marker='o', linestyle='--', label="Jacobi")

    # Gauss-Seidel
    plt.plot(TG[0, :], TG[1, :], marker='s', linestyle='-', label="Gauss-Seidel")

    # Punto inicial x0
    plt.scatter(TJ[0, 0], TJ[1, 0], marker='*', s=200, label="$x^{(0)}$")

    # Solución final
    plt.scatter(TJ[0, -1], TJ[1, -1], marker='X', s=120, label="Solución")

    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.title("Trayectoria de Jacobi vs Gauss-Seidel")
    plt.legend()
    plt.grid(True)
    plt.axis("equal")
    plt.show()
```

```python
A = [[1,1],[-2,5]]
b = [[7],[0]]
n = 2
x_j, traj_j = gauss_jacobi(A=A, b=b, x=[0] * n, tol=1e-5, max_iter=10)  # type: ignore

print(f"Valores de x:\n {x_j}")
print(f"Valores de las iteraciones:\n {traj_j}")
```

```python
A = [[1,1],[-2,5]]
b = [[7],[0]]
n = 2
x_j2, traj_j2 = gauss_seidel(A=A, b=b, x=[0] * n, tol=1e-5, max_iter=10)  # type: ignore
print(f"Valores de x:\n {x_j2}")
print(f"Valores de las iteraciones:\n {traj_j2}")
```

```python
graficar_trayectorias_2D(traj_j,traj_j2)
```
